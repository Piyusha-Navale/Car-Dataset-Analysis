```python
plt.figure(figsize=(12, 8))
sns.countplot(x='Owner_Count', data=data, order=data['Owner_Count'].value_counts().index[:10])
plt.xticks(rotation=45)
plt.title('Count of the orders')
plt.show()
```

```python

# Data for pie chart
labels = ['Year', 'Mileage', 'Fuel_Type', 'Owner_Count','Price']
sizes = [20, 30, 25, 25,20]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','Pink']

# Create a pie chart
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Show the pie chart
plt.title('Pie Chart Example')
plt.show()
```

```python


# Data for clustered bar chart
labels = ['Engine_Size', 'Owner_Count', 'Year', 'Price']
data1 = [20, 35, 30, 35]
data2 = [25, 32, 34, 20]

# Set up the positions for bars
x = np.arange(len(labels))
width = 0.35  # width of the bars

# Create a clustered bar chart
fig, ax = plt.subplots(figsize=(8, 6))

# Create bars for both datasets
bars1 = ax.bar(x - width/2, data1, width, label='Data1')
bars2 = ax.bar(x + width/2, data2, width, label='Data2')

# Add labels, title, and custom x-axis tick labels
ax.set_xlabel('Category')
ax.set_ylabel('Values')
ax.set_title('Clustered Bar Chart Example')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Show the chart
plt.show()
```

```python

# Data for histogram
data = np.random.randn(1000)

# Create a histogram
plt.hist(data, bins=30, edgecolor='black')

# Add labels and title
plt.xlabel('Engine_Size')
plt.ylabel('Mileage')
plt.title('Histogram Example')

# Show the chart
plt.show()
```
